**Exercise 1.1.** Assume the one-period binomial market of Section 1.1 that both $H$ and $T$ have positive probability of occurring. Show that condition (1.1.2) precludes arbitrage. In other words, show that if $X_0 = 0$ and

$$X_1 = \Delta_0 S_1 + (1 + r) (X_0 - \Delta_0 S_0),$$

then we cannot have $X_1$ strictly positive with positive probability unless $X_1$ is strictly negative with positive
probability as well, and this is the case regardless of the choice of the number $\Delta_0$.

---

The excercise asks us to prove that

$$
\begin{cases}
X_1(H) > 0 \Rightarrow X_1(T) < 0 \\
X_1(H) < 0 \Rightarrow X_1(T) > 0
\end{cases},
$$

$$\forall \Delta_0 \in \mathbb{R}.$$

---

As given by the excercise:

$$X_0 = 0.$$

Then

$$X_1 = \Delta_0 S_1 - (1 + r) \Delta_0 S_0.$$

We know that

$$
\begin{cases}
S_1 = S_0 u \\
S_1 = S_0 d
\end{cases}.
$$

Therefore

$$
\begin{cases}
    X_1(H) = \Delta_0 S_0 (u - (1 + r)) \\
    X_1(T) = \Delta_0 S_0 (d - (1 + r))
\end{cases}.
$$

Given condition (1.1.2):

$$d < 1 + r < u,$$

we have:

$$
\begin{cases}
u - (1 + r) > 0 \\
d - (1 + r) < 0
\end{cases}.
$$

By analysing the products for $X_1(H)$

$$\Delta_0 S_0 (u - (1 + r))$$

and for $X_1(T)$

$$\Delta_0 S_0 (d - (1 + r)),$$

the following propositions hold:

$$
\begin{cases}
\Delta_0 < 0 \land S_0 < 0 \Rightarrow X_1(H) > 0 \land X_1(T) < 0 \\
\Delta_0 < 0 \land S_0 > 0 \Rightarrow X_1(H) < 0 \land X_1(T) > 0 \\
\Delta_0 > 0 \land S_0 < 0 \Rightarrow X_1(H) < 0 \land X_1(T) > 0 \\
\Delta_0 > 0 \land S_0 > 0 \Rightarrow X_1(H) > 0 \land X_1(T) < 0 \\
\Delta_0 = 0 \lor S_0 = 0 \Rightarrow X_1(H) = X_1(T) = 0
\end{cases}.
$$

With all the possible values of $\Delta_0$ and $S_0$ covered, we conclude that, for any case,

$$
\begin{cases}
X_1(H) > 0 \Rightarrow X_1(T) < 0 \\
X_1(H) < 0 \Rightarrow X_1(T) > 0
\end{cases},
$$

$$\forall \Delta_0 \in \mathbb{R}.$$

$$\square$$

Alternatively, definig

$$a = sgn(X_1)$$

$$b = sgn(\Delta_0),$$

$$c = sgn(S_0)$$

and

$$
d = \begin{cases}
sgn(u - (1 + r)) = 1 \; if \; X_1 = X_1(H) \\
sgn(d - (1 + r)) = -1 \; if \; X_1 = X_1(T)
\end{cases},
$$

where $sgn$ is the sign function.

We have

$$sgn(X_1) = sgn(\Delta_0 S_1 - (1 + r) \Delta_0 S_0),$$

which is equivalent to

$$
sgn(X_1) =
\begin{cases}
    sgn(X_1(H)) = sgn(\Delta_0 S_0 (u - (1 + r))) \\
    sgn(X_1(T)) = sgn(\Delta_0 S_0 (d - (1 + r)))
\end{cases}.
$$

Since

$$sgn\left(\prod_{i = 1}^n x_i\right) = \prod_{i = 1}^n sgn(x_i),$$

then

$$
sgn(X_1) =
\begin{cases}
    sgn(X_1(H)) = sgn(\Delta_0) \cdot sgn(S_0) \cdot sgn(u - (1 + r))) \\
    sgn(X_1(T)) = sgn(\Delta_0) \cdot sgn(S_0) \cdot sgn(d - (1 + r)))
\end{cases},
$$

so

$$sgn(X_1) = a = b \cdot c \cdot d.$$

Therefore,

```julia
# image of sgn function
Y = (-1, 0, 1)
for b ∈ Y, c ∈ Y
    if b == -1
        print("Δ₀ < 0 ∧ ")
    elseif b == 0
        print("Δ₀ = 0 ∧ ")
    else
        print("Δ₀ > 0 ∧ ")
    end
    if c == -1
        print("S₀ < 0 ⇒ ")
    elseif c == 0
        print("S₀ = 0 ⇒ ")
    else
        print("S₀ > 0 ⇒ ")
    end
    # X₁ = X₁(H)
    d = 1
    a = b*c*d
    if a == -1
        print("X(H) < 0 ∧ ")
    elseif a == 0
        print("X(H) = 0 ∧ ")
    else
        print("X(H) > 0 ∧ ")
    end
    # X₁ = X₁(T) 
    d = -1
    a = b*c*d
    if a == -1
        println("X(T) < 0")
    elseif a == 0
        println("X(T) = 0")
    else
        println("X(T) > 0")
    end
end 
```
